```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/prabh/OneDrive/Desktop/Summer_2025/Fake_News_detection_using_NLP/dataset/Fake/Fake.csv')
df2 = pd.read_csv('C:/Users/prabh/OneDrive/Desktop/Summer_2025/Fake_News_detection_using_NLP/dataset/True/True.csv')
df1["Reasult"] = 0
df2["Reasult"] = 1
df1.drop(['subject'],axis=1,inplace=True)
df2.drop(['subject'],axis=1,inplace=True)
concatenated_df = pd.concat([df1, df2], ignore_index=True)

```

```python
concatenated_df["combined"] = concatenated_df['title']+" "+concatenated_df["text"]
```

##Text PreProcessing Pipeline 

==>Text to lower case 

```python
concatenated_df["combined"]=concatenated_df["combined"].str.lower()
```

```python
concatenated_df
```

==> Removing HTML tags


```python
import re 
def remove_html_tags(text):
    pattern=re.compile('<.*?')
    return pattern.sub(r'',text)
```

```python
concatenated_df['combined']=concatenated_df['combined'].apply(remove_html_tags)
```

==>Removing URL

```python
import re

def remove_urls(text):
    """Remove URLs from a text string"""
    return re.sub(r'http[s]?://\S+', '', text)
```

```python
concatenated_df['combined']=concatenated_df['combined'].apply(remove_urls)
```

==>Removing Punctuation Marks


```python
import string
exclude=string.punctuation
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
concatenated_df['combined']=concatenated_df['combined'].apply(remove_punc)
```
